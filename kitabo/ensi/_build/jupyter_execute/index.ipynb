The ghost in "Hamlet" is indeed fundamentally different from typical fantasy elements, and that's a crucial point I overlooked earlier. Let's break this down:

1. Lack of agency: As you noted, Hamlet's father's ghost doesn't have independent agency. It appears, delivers information, and disappears. It doesn't actively participate in events beyond its initial revelation.

2. Plot device vs. character: The ghost serves primarily as a catalyst for the plot, not as an ongoing active character. This is distinct from many fantasy works where supernatural entities are fully realized characters with their own arcs and agency.

3. Ambiguity: There's deliberate ambiguity about whether the ghost is "real" or a product of Hamlet's mind, which is less common in traditional fantasy where supernatural elements are usually unambiguously real within the story's world.

4. Limited interaction: The ghost interacts only with Hamlet and his friends initially, and then only with Hamlet. It doesn't engage with the broader world of the play.

5. Thematic function: The ghost serves more as a representation of the past and unresolved issues than as a fantastical element for its own sake.

Given these points, classifying "Hamlet" as fantasy based solely on the ghost's presence would indeed be a stretch. The ghost functions more as a dramatic device to set the plot in motion and explore themes of revenge, duty, and the weight of the past.

This analysis aligns more closely with your comparison to "The Canterville Ghost," where the supernatural element is more a vehicle for the story's themes and plot than a defining genre feature.

In reconsidering, while "Hamlet" has a supernatural element, its function within the play doesn't align with the role supernatural elements typically play in the fantasy genre. The ghost is a means to an end - a way to reveal truth and motivate action - rather than an end in itself or a window into a broader fantastical world.

````python
import numpy as np

# Create a simple dataset (design matrix: collective unconscious)
X = np.array([
    [1, 2],
    [1, 3],
    [1, 4],
    [1, 5],
    [1, 6]
])

y = np.array([2, 3, 4, 4, 5])

# Calculate beta coefficients
X_transpose = X.T
X_transpose_X = np.dot(X_transpose, X)
X_transpose_y = np.dot(X_transpose, y)
beta = np.linalg.inv(X_transpose_X).dot(X_transpose_y)

print("Beta coefficients:")
print(beta)

# Predict y values
y_pred = np.dot(X, beta)

print("\nPredicted y values:")
print(y_pred)

print("\nActual y values:")
print(y)
````
